Jeg valgte moving average som filtreringsalgoritme fordi den er enkel å implementere i python, samtidig som den reduserer tilfeldige variasjoner i dataene ved å erstatte hver verdi med gjennomsnittet av et fast antall nærliggende observasjoner. Moving average er god på å jevne ut raske svingninger i akselerasjonsmålinger, men man på passe på at det kan redusere skarpheten i signalet dersom antall observasjoner er for stort. 

4.	Kildekode: Python-kode som viser hvordan du filtrerte bort støyen.

Akselrasjenen i intervallet 7 til 10 sekunder har en relativt stabil akselerasjon. Derfor passer det bra å se på støyredusering i dette intervallet.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("Raw Data.csv", delimiter=";", decimal=".")

def moving_average(signal, window_size):
    return np.convolve(signal, np.ones(window_size)/window_size, mode='valid')

window_size = 10
smoothed_acceleration = moving_average(data["Acceleration z (m/s^2)"].values, window_size)
time_smoothed = data["Time (s)"].values[:len(smoothed_acceleration)]

plt.scatter(data["Time (s)"], data["Acceleration z (m/s^2)"], color='black', marker='o', label='Rådata (z-retning)', s=10, alpha=0.5)
plt.plot(time_smoothed, smoothed_acceleration, color='red', label='Filtrert data', linewidth=2)

plt.xlim([7,10])
plt.xlabel('Tid [sekunder]')
plt.ylabel('Akselerasjon [m/s**2]')
plt.title('Akselerasjonsmålingene med filtrering')
plt.legend()

plt.show()
```

# 7. Beregne farten fra akselerasjonsdata

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import cumulative_trapezoid as cumtrapz

filnavn = 'Raw Data_1.csv'
data = np.genfromtxt(filnavn, delimiter=';', skip_header=1)

def numerisk_integrasjon(t, a, v0=0, s0=0):
    v = cumtrapz(a, t, initial=v0)
    s = cumtrapz(v, t, initial=s0)
    return v, s


t = data[:, 0]
a_x = data[:, 1]
a_y = data[:, 2]
a_z = data[:, 3]

v0 = 0
s0 = 0

v_z, s_z = numerisk_integrasjon(t, a_z, v0, s0)

fig, axs = plt.subplots(2, 1, figsize=(8, 10))


axs[0].plot(t, v_z, label='Fart (m/s) i z-retning', color='blue')
axs[0].set_ylabel('Fart (m/s)')
axs[0].set_xlabel('Tid (s)')
axs[0].set_title('Heisens fart')
axs[0].grid()
axs[0].legend()

axs[1].plot(t, s_z, label='Posisjon (m) i z-retning', color='green')
axs[1].set_xlabel('Tid (s)')
axs[1].set_ylabel('Posisjon (m)')
axs[1].set_title('Heisens posisjon')
axs[1].grid()
axs[1].legend()

plt.tight_layout()
plt.show()
```

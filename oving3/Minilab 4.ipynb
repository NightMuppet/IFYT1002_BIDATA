```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import cumulative_trapezoid as cumtrapz

filnavn = 'Raw Data_1.csv'
data = np.genfromtxt(filnavn, delimiter=';', skip_header=1)

def numerisk_integrasjon(t, a, v0=0, s0=0):
    """
    Beregner fart og posisjon fra akselerasjonsdata ved hjelp av trapesmetoden.
    """
    v = cumtrapz(a, t, initial=v0)
    s = cumtrapz(v, t, initial=s0)
    return v, s


t = data[:, 0]
a_x = data[:, 1]
a_y = data[:, 2]
a_z = data[:, 3]

v0 = 0
s0 = 0

v_z, s_z = numerisk_integrasjon(t, a_z, v0, s0)

fig, axs = plt.subplots(2, 1, figsize=(8, 10))


axs[0].plot(t, v_z, label='Fart (m/s) i z-retning', color='blue')
axs[0].set_ylabel('Fart (m/s)')
axs[0].set_xlabel('Tid (s)')
axs[0].set_title('Heisens fart')
axs[0].grid()
axs[0].legend()

axs[1].plot(t, s_z, label='Posisjon (m) i z-retning', color='green')
axs[1].set_xlabel('Tid (s)')
axs[1].set_ylabel('Posisjon (m)')
axs[1].set_title('Heisens posisjon')
axs[1].grid()
axs[1].legend()

plt.tight_layout()
plt.show()
```

### Refleksjon og feilsking

1. Verdiene som blir plotet er realistiske, heiser beveger seg ofte i en hastighet p mellom 0.5 -2 m/s.

2. Heisen har hastilghet lik null der grafen for posisjon/tid er flat. Dette er i intervallene fra 0 til 4 sekunder, og fra 14 til 18 sekunder

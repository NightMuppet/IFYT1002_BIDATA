# 5. Kunne plukke ut et selectivt utvalg fra dataseriene

For å finne maksimal akselerasjon kan vi plukke ut og data plotte for z-koordinaten for tidspunkt mellom 1 og 6 sekunder 

Her ser jeg at området rundt 7 til 10 ser det ut som om vi har en relativt stabil akselerasjon. Nå har jeg behov for å plukke ut målingene som ligger i dette intervallet

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Laste inn data fra CSV-filen med riktig separator
data = pd.read_csv("Raw Data.csv", delimiter=";", decimal=".")

# Definere glidende gjennomsnitt funksjon
def moving_average(signal, window_size):
    return np.convolve(signal, np.ones(window_size)/window_size, mode='valid')

window_size = 10  # Juster etter behov
smoothed_acceleration = moving_average(data["Acceleration z (m/s^2)"].values, window_size)
time_smoothed = data["Time (s)"].values[:len(smoothed_acceleration)]  # Matche tidsverdier

plt.scatter(data["Time (s)"], data["Acceleration z (m/s^2)"], color='black', marker='o', label='Rådata (z-retning)', s=10, alpha=0.5)
plt.plot(time_smoothed, smoothed_acceleration, color='red', label='Filtrert data', linewidth=2)

plt.xlim([7,10])
plt.xlabel('Tid [sekunder]')
plt.ylabel('Akselerasjon [m/s**2]')
plt.title('Akselerasjonsmålingene med filtrering')
plt.legend()

plt.show()
```

Gjør nå tilsvarende for å finne den minste akselerasjonen jeg er utsatt for. 

```python
# Søker opp intervallet mellom 5.4 til 6.3 sekunder
intervallB = (data[:,0] >= 8.2) & (data[:,0] <= 9) 
plt.scatter(data[intervallB,0] , data[intervallB,3] , color='green', marker='o', label='Data points')
#plt.ylim([8.5,11])
# Add labels and a title
plt.xlabel('Tid [sekunder]')
plt.ylabel('Akselerasjon [m/s**2]')
plt.title('Akselerasjonsmålingene')
 
plt.show()
```

```python
# Regner ut middelverdi og standardavviket for målingene i intervallet. 

print(f'Maksimal akselerasjon {np.mean(data[intervallB,3]):.4g} ± {np.std(data[intervallB,3]):.2g}')
```

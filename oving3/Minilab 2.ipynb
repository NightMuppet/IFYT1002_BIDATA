# 3 Importere fila i jupyter / python 

Nå er det bare å lagre fila og pakke den ut på et sted der du også kan lagre denne jupyter notebook fila. I eksemplet under henter jeg ut data fra en kjøring fra E1 til E4 i Realfagbygget bokk E 20 januar 2025. Først må vi se på dataene. Under har vi hentet inn alle målingne og plottet sammen. 

```python
import numpy as np
import matplotlib.pyplot as plt
# Replace 'your_file.csv' with the path to your CSV file
#filnavn='C:\\Users\\trotho\\OD\\fag\\IFYT1001\\programmering\\øvinger\\Raw Data_u1_5.csv' # ligger på et annet sted
filnavn='Raw Data_1.csv' # ligger samme sted som jupyter notebook fila
data = np.genfromtxt(filnavn, delimiter=';', skip_header=1)  # Use skip_header=1 to skip the header row if there is one
 
#print(data)
 
plt.scatter(data[:,0] , data[:,1] , color='blue', marker='o', label='Data points')
plt.scatter(data[:,0] , data[:,2] , color='red', marker='o', label='Data points')
plt.scatter(data[:,0] , data[:,3] , color='green', marker='o', label='Data points')
#plt.xlim([8,20])
#plt.ylim([8.5,11])
# Add labels and a title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Scatterplot')
 
# Add a legend
plt.legend()
 
# Show the plot
plt.show()
```

Her er målingen gjort med tyngdekraften i bildet. For å kunne se på dataene må vi se nærmere på noen helt spesifikke tidspunkter. Begrenser plottet til å bare ha med den grønne kurven og ser bare på tidspunkt mellom 20 og 30 sekunder

Her ser jeg at området rundt 22.50 til 23.25 ser det ut som om vi har en stabil akselerasjon. Nå har jeg behov for å plukke målingene som ligger i et intervall

```python
intervall = (data[:,0] >= 22.50) & (data[:,0] <= 23.25)
print(intervall)
```

```python
intervall = (data[:,0] >= 22.5) & (data[:,0] <= 23.25)
plt.scatter(data[intervall,0] , data[intervall,3] , color='green', marker='o', label='Data points')
#plt.ylim([8.5,11])
# Add labels and a title
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Simple Scatterplot')
 
# Add a legend
plt.legend()
 
# Show the plot
plt.show()
```
